```python
import numpy as np
import matplotlib.pyplot as plt
import time
import timeit
```

```python
#Problem 2
def while_M():
    L = 100

    M = 0.

    i = -L
    while i <= L:
        j = -L
        while j <= L:
            k = -L
            while k <= L:
                if (i == 0) * (j == 0) * (k == 0): #origin 
                    k += 1
                else:
                    num = (-1)**(i + j + k) #going out from origin
                    M += float(num)/(i**2+j**2+k**2)**0.5 
                    k += 1
            j += 1
        i += 1
    print(-M)
```

```python
def for_M():
    L = 100

    M = 0.

    i = -L
    for i in range(-L, L):
        j = -L
        for j in range(-L, L):
            k = -L
            for k in range(-L, L):
                if (i == 0) * (j == 0) * (k == 0): #origin 
                    continue
                else:
                    num = (-1)**(i + j + k) #going out from origin
                    M += float(num)/(i**2+j**2+k**2)**0.5 
    print(-M)
```

```python
timeit.timeit('while_M()', 'from __main__ import while_M', number=1)
```

```python
timeit.timeit('for_M()', 'from __main__ import for_M', number=1)
```

```python
#Problem 3
N = 1000 
numIterations = 100
grid = np.full([ N, N ], numIterations, int) 

cx, cy = np.meshgrid(linspace(-2, 2, N), linspace(-2, 2, N)) 

z = zeros([N,N], complex)

for n in range(0, numIterations):
    z = z**2 + cx + cy*1j
    for i in range(0, N):
        for j in range(0, N):
            if abs(z[i,j]) > 2:
                grid[i,j] = n+1

fig = plt.figure(figsize=(8, 8), dpi=150)                 
plt.imshow(grid,origin='lower',extent=[-2,2,-2,2])
plt.gray()
plt.show()
```

```python
#Problem 4
def quadratic(a, b, c): 
    # calculate the discriminant
    d = (b**2) - (4*a*c)

    # find two solutions for each part
    sol1 = (-b - ((d)**0.5))/(2*a)
    sol2 = (-b + ((d)**0.5))/(2*a)
    print("The two solutions for part (a) are: ", sol1 ,"and ", sol2)
    
    sol3 = ((2*c)/(-b + (d**0.5)))
    sol4 = ((2*c)/(-b - (d**0.5)))
    print("The two solutions for part (b) are: ", sol3 ,"and ", sol4)
    
    print("The two solutions for part (c) are: ", sol1, "and ", sol4)
    
quadratic(.001,1000,.001)
```
