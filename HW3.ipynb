```python
#Problem 1
#part a
def f(x):
    return x * (x-1)
    
delta = 10**-2;
#for x=1
term = (f(1+delta) - f(1))/delta

print (term)
print ("True Value = 1" )
    
#part b
delta2 = [10**-4, 10**-6, 10**-8, 10**-10, 10**-12, 10**-14];
for delta in delta2:
    term = (f(1+delta) - f(1))/delta
    print(term)
    
#the error reduces at first because as h decreases,
#the trunction error decreases, however rounding error 
#increases with decrease in h, as a result the error 
#increases after a while
```

```python
#Problem 2

from numpy import zeros, ones
import time

def matrix(N):
    A=ones([N,N], float)
    B=ones([N,N], float)
    C=zeros([N,N],float)

    for i in range(N):
        for j in range(N):
            for k in range(N):
                C[i,j]+=A[i,k]*B[k,j]
    return C

all_N=[10,30,50,70,90,110]
list =[];
for i in all_N:
    start = time.time()
    matrix(i)
    end = time.time()
    list.append(end - start)
    
plt.plot(all_N, list)
plt.title('Time VS Matrix Size')
plt.xlabel('Matrix Size')
plt.ylabel('Time (s)')
plt.show()
plt.title('log(Time) VS log(Matrix Size)')
plt.xlabel('log(Matrix Size)')
plt.ylabel('log(Time (s))')
plt.loglog(all_N, list)
slope, intercept = np.polyfit(np.log(all_N), np.log(list), 1)
print('slope of loglog = ', slope)
plt.show()
#does not rise by N^3, rises by a little less than that
```

```python
#Problem 3
import matplotlib.pyplot as plt
import numpy as np

nwalkers = 10000
nsteps = 1000
steps = np.random.random(size=(nwalkers, nsteps)) - 0.5 

paths = steps.cumsum(axis=1)
for indx in np.arange(nwalkers):
    plt.plot(paths[indx, :], alpha=0.2)
plt.xlabel('$N_{step}$')
plt.ylabel('x')
```

```python
means = np.mean(paths, axis=0)
plt.plot(means)
```

```python
stds = np.std(paths, axis=0)
plt.plot(stds)
```
