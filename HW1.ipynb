```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats

mean=0.;
std=3.;
x_min=-10.;
x_max=10.;
x = np.linspace(x_min, x_max, 100)
y = scipy.stats.norm.pdf(x,mean,std)
plt.plot(x,y, color = 'pink')
plt.title("Gaussian Distribution")
plt.xlabel("x")
plt.ylabel("Distribution")
plt.show()
```
